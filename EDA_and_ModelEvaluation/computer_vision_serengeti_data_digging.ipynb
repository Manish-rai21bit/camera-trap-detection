```python
data_path = '../data/Computer_Vision_Serengeti/'
```

```python
gold_standard_bb = {}
data ={}
with open((data_path + 'computer-vision-serengeti-subjects.csv'), 'r') as csvfile:
    csvdata = csv.reader(csvfile, delimiter=',')
    header = next(csvdata)
    col_mapper = {k: i for i, k in enumerate(header)}
    for i, row in enumerate(csvdata):
        filename = row[col_mapper['subject_id']]
#         if filename not in gold_standard_bb.keys():
#             gold_standard_bb[filename] = []
#         gold_standard_bb[filename].append({k: row[v] for k, v in col_mapper.items() })
#         print(list(json.loads(row[4]).keys()))
        if 'Filename' in list(json.loads(row[4]).keys()):
            data[filename] = [json.loads(row[4])['Filename'], '',json.loads(row[5])['0']]
        else:
            data[filename] = [json.loads(row[4])['Subject ID'], json.loads(row[4])['Crowd Determination'], 
                              json.loads(row[5])['0']]

#         print(data)
        if i%5000 == 0:
            print("performing iteration : %s" % i)
```

```python
df2 = pd.DataFrame.from_dict(data, orient='Index', columns=['Subject_id', 'Species', 'URL'])
df2.head()
```

```python
df3 = df2.groupby(by=['Species'], as_index=False).count()
df3 = df3[['Species', 'URL']]

df2_freq = df3.sort_values(by='URL')\
                            .rename(index=str, columns={'URL':'freq'})

# df_bb_species_freq['bbox_freq_per_image'] = round((df_bb_species_freq['bbox_freq']/df_images.shape[0])*100,2)

plt.figure(figsize=(8,10))
plt.grid(True)
plt.barh(df2_freq['Species'], df2_freq['freq'], align='center', alpha=0.75)
plt.title('Frequency Distribution of images with species in Computer Vision: Serengeti')
plt.xlabel('Number of bounding box in a total {0} images'.format(df2.shape[0]))
```
