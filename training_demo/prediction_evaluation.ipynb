### Confusion matrix


```python
df = df_comparison

df.loc[df["species_count_prediction"]<=df["species_count_truth"], "true_positive"] = df["species_count_prediction"]
df.loc[df["species_count_prediction"]>df["species_count_truth"], "true_positive"] = df["species_count_truth"]

df.loc[df["species_count_prediction"]>df["species_count_truth"], "false_positive"] = df["species_count_prediction"] - df["species_count_truth"]
df.loc[df["species_count_truth"]>df["species_count_prediction"], "false_negative"] = df["species_count_truth"] - df["species_count_prediction"]

df.loc[df["class_name"] == "Wildebeest", "two_class_species"] = df["class_name"]
df.loc[df["class_name"] != "Wildebeest", "two_class_species"] = "not_" + "Wildebeest"
```

```python
df = df.fillna(0)
df.head(n = 10)
```

```python
# Accuracy, Precision and Recall at Species level
?df.groupby
```

```python
df1 = df.groupby(by=['class_name'], axis=0, as_index=False)["species_count_truth", "species_count_prediction", \
                                                            "true_positive", "false_positive", "false_negative"].sum()
df1.head()
```

```python
df1["precision"] = (df1["true_positive"]/(df1["true_positive"] + df1["false_positive"]))*100
df1["accuracy"] = (df1["true_positive"]/(df1["true_positive"] + df1["false_negative"]))*100
df1 = df1.fillna(0)

# rounding accuracy and precision
df1.precision = df1.precision.round(1)
df1.accuracy = df1.accuracy.round(1)

#sorting by values of precision and accuracy
df1 = df1.sort(columns=["species_count_truth", "precision", "accuracy"], axis=0, ascending=False)\
        .reset_index(drop=True)
df1["recall"] = df1["accuracy"]
```
